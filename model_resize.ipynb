## Model

```python
model = Wav2Vec2ForCTC.from_pretrained(
        "facebook/wav2vec2-base-960h"
    )
```

```python
model.lm_head = nn.Linear(in_features=768, out_features=58, bias=True)
```

```python
model.config.vocab_size = 58
```

```python
model.save_pretrained("model_korean_base")
```
